<h1 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
4. Critical Evaluation and Discussion
</h1>

<h2 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
(a) Agreement/Disagreement with Findings:
</h2>

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Agreement on Model Improvement:
</h3>
Our results support the authors' conclusion that the Five-Factor model outperforms the Three-Factor model. In the original paper, Fama and French identify the "microcap extreme growth" portfolio as a major failure of the three-factor model. In our replication of the Three-Factor model on the 25 Size-B/M portfolios, the "Small Low B/M" portfolio produced a  negative intercept (α) of -0.742% per month with a t-statistic of -4.78. When we ran the Five-Factor model on the same data, this intercept shrank to -0.276% (t-statistic -3.25). This confirms the authors' finding that adding Profitability (RMW) and Investment (CMA) helps explain—though does not fully eliminate—the pricing errors of small growth stocks.


Though our replication supports the authors’ main conclusions, our numerical estimates do not match theirs exactly. The key economic patterns emphasized in the paper—such as strong size, value, profitability, and investment effects, as well as the characteristic loading patterns across portfolios—are clearly present in our results. For example, small firms consistently exhibit higher SMB exposure, high-B/M portfolios load strongly on HML, and firms with robust profitability or conservative investment policies show the expected positive RMW and CMA loadings. These qualitative relationships align closely with the original study. However, the exact regression coefficients and 
t-statistics differ from the published tables, which we attribute to differences in the scale of our recplication relative to the original paper's, rather than inconsistencies in the model itself. As a result, we do not question the authors’ conclusions; instead, our evidence suggests that the five-factor model behaves consistently across datasets, even though numerical estimates vary.

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Confirmation of "Redundancy":
</h3>
We agree with the finding that HML is largely redundant in the presence of RMW and CMA. The authors state that "the average HML return is captured by the exposures of HML to other factors". Our replication provides the statistical mechanism for this: we found a correlation of 0.70 between HML and CMA. Furthermore, in our sample statistics, CMA had a higher t-statistic (4.01) than HML (3.38), suggesting that the investment factor is a more robust driver of returns than the value factor in this sample period.

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Agreement on Rejection (With Discrepency in GRS):
</h3>
We agree with the authors that the model is statistically incomplete, but while we agree the model is imperfect, we question the uniformity of its failure. The authors state the GRS test "easily rejects" the model. However, our replication reveals a massive disparity in how strongly it is rejected. For the 25 Size-B/M portfolios, we calculated a GRS statistic of 63.20 (p≈0). In contrast, for the 25 Size-OP portfolios, the GRS statistic was drastically lower at 1.89 (p=0.005). This suggests that the model's failure is not general, but is specifically acute when assets are sorted by Book-to-Market ratios, a nuance that perhaps warrants more emphasis than a blanket rejection.

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Confirmation of Correlations:
</h3>
We agree with the structural relationships between factors. We replicated the correlation matrix exactly, finding a correlation of 0.70 between HML and CMA and -0.34 between SMB and RMW. This supports the economic logic that value firms invest conservatively, and small firms tend to be less profitable.

<h2 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
(b) Economic Interpretation:
</h2>

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Magnitudes and Valuation Theory:
</h3>
The magnitudes of the premiums we estimated make economic sense under the Dividend Discount Model ($M_t=∑E(Y−dB)/(1+r)$). We found a mean monthly premium of 0.26% for RMW and 0.33% for CMA. Economically, this implies that investors require a higher return (discount rate $r$) to hold firms that have high expected earnings ($Y$) or low expected asset growth ($dB$), holding market value ($M$) constant. The positive coefficients on RMW for high-profitability portfolios in our regressions (e.g., 1.07 for Small HiOP)  confirm that the market rationally prices profitability as a risk factor.

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Policy Implications:
</h3>
The evidence supports an investment policy that favors "Quality" (Profitability/Investment) over traditional "Value." In our regression results, the HML factor had a lower mean return (0.38%) compared to the combined effect of RMW and CMA. Additionally, the HML factor loadings were often subsumed by CMA loadings in the regressions for high B/M portfolios. For asset managers, this implies that "Value" strategies might just be noisy proxies for "Conservative Investment" strategies. Shifting policy to target CMA directly could provide similar returns with potentially less volatility (CMA Std Dev 2.00 vs HML Std Dev 2.79)

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Limitations for External Validity:
</h3>
A major limitation is the sample period cutoff. Our replication strictly followed the authors' cutoff of December 2013. Economic conditions have changed significantly since 2013 (e.g., near-zero interest rates, the rise of intangible assets). The validity of these factor premiums in the post-2013 economy remains an open question in this specific analysis. The post-2013 economy has been characterized by the dominance of large-cap technology firms (high profitability, high investment). The model's "lethal" failure to price small, high investment firms (which we confirmed with a -0.36% alpha for Small HiInv) might be exacerbated in a modern economy driven by speculative tech growth, limiting the model's validity for current portfolios.

<h2 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
(c) Suggestions for Improvement:
</h2>

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Investigation of the GRS Discrepancy:
</h3>
A robust check should investigate why the GRS statistic for the Size-B/M portfolios (63.20) is drastically higher than for the Size-Inv portfolios (3.52) or Size-OP portfolios (1.89). This huge disparity suggests that the standard "Size-B/M" sort creates specific linear combinations of assets that the Five-Factor model is particularly bad at pricing, which warrants deeper investigation than just reporting the rejection.

The authors identify small, low-profitability, high-investment stocks as the main problem. To strengthen the paper, we suggest a "trimmed" robustness check. If we exclude the bottom decile of "microcap growth" stocks (which are often illiquid and hard to short), does the model pass the GRS test? Our analysis showed the Small Low B/M alpha was the largest negative outlier (-0.28%). Removing these hard-to-trade assets might reveal that the model works perfectly well for the investable universe.

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Transaction Cost Analysis:
</h3> 
The paper assumes frictionless trading. However, the SMB factor (which we found has a mean return of 0.28%) relies on long positions in small-cap stocks. A robustness check should calculate factor returns net of bid-ask spreads. If the "Size" premium disappears after transaction costs, the five-factor model might be overfitting to liquidity premiums rather than fundamental economic risk.

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Alternative Factor Construction:
</h3>
The authors use 2×3 sorts to build factors. We suggest testing if factors built from separate univariate sorts perform better. The high correlation we found between HML and CMA (0.70)  suggests that the sorting methodology itself might be inducing multicollinearity. Constructing orthogonal factors could help isolate the true marginal contribution of Investment versus Value.

<h1 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
5. Conclusion and Reflection
</h1>

<h2 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
(a) Summary: 
</h2>

We successfully replicated the Fama-French (2015) Five-Factor model using monthly data from July 1963 to December 2013 (N=606). Our analysis confirms the authors' descriptive results. For instance, our replication found a mean monthly return for the Market factor ($R_M−R_F$) of 0.50% and Size factor (SMB) of 0.28%, which closely matches the original paper’s reported means of 0.50% and 0.29%. Our regression analysis confirmed that the Five-Factor model reduces the unexplained average returns (alpha) for small growth stocks compared to the Three-Factor model. The intercept (α) for the Small Low B/M portfolio improved from -0.74% (t=−4.78) in the Three-Factor specification to -0.28% (t=−3.25) in the Five-Factor specification. However, GRS tests on six different portfolio sets confirmed that the model is still statistically rejected (p<0.01 in all cases).

Crucially, we replicated the structural relationships between factors that drive the paper's main conclusions. We calculated a correlation of 0.70 between the Value (HML) and Investment (CMA) factors , an exact match to the 0.70 correlation reported by Fama and French. This high correlation statistically supports the paper's finding that HML is "redundant" because its information is subsumed by the investment factor. Furthermore, our regression analysis confirmed that the Five-Factor model significantly reduces pricing errors for small growth stocks.

<h2 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
(b) Overall Assessment: 
</h2>

This paper represents a significant theoretical advancement in asset pricing. Unlike the Three-Factor model, which was largely empirically motivated, the Five-Factor model is rigorously grounded in the Dividend Discount Model. The paper successfully bridges the gap between empirical anomalies and valuation theory. By demonstrating that HML is largely redundant due to its correlation with CMA, the authors provide a more parsimonious view of market risks. However, the persistent negative alpha for small, high-investment, low-profitability firms (the "lethal" portfolio) remains a significant unresolved issue, suggesting the model is still missing a factor related to speculative growth.

As the authors state, the valuation equation implies that "higher expected earnings imply a higher expected return" and "higher expected growth in book equity investment implies a lower expected return".

Our replication confirms that this theoretical insight holds empirically: we found robust positive premiums for Profitability (RMW, mean 0.26%) and Investment (CMA, mean 0.33%). However, we assess the model as "incomplete" based on our GRS test results. We calculated a GRS statistic of 63.20 (p≈0) for the 25 Size-B/M portfolios , confirming the authors' admission that the GRS test "easily rejects" the model. Specifically, our finding of a persistent -0.36% monthly alpha for Small High-Investment stocks validates the authors' conclusion that the model suffers from a "lethal" inability to price small firms that invest aggressively despite low profitability.

<h2 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
(c) Learning Reflection: 
</h2>

This replication process provided a practical lesson in the sensitivity of econometric research to data vintage. Although we restricted our sample to the exact 606 months used by the authors, we observed slight discrepancies in volatility metrics—for example, our sample's HML standard deviation was 2.79% compared to the paper's 2.88%. This taught us that financial databases like CRSP and Compustat are not static; historical data is often "restated" or corrected years later.

Additionally, implementing the GRS test deepened our understanding of multivariate testing. We learned that the test's power depends heavily on the precision of the residual covariance matrix (Σ). The massive disparity we found between the GRS statistic for Size-B/M portfolios (63.20) and Size-OP portfolios (1.89)  highlights how sensitive asset pricing models are to the specific characteristics used to sort test assets.

<h2 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
(d) Transparency in Research: 
</h2>

This project highlighted how critical data transparency is. This was seen as we could download the exact csv files for factors and portfolios and match the authors' exact sample count of 606 months, we could distinguish between "failed replication" and "statistical noise." Without publicly available data libraries like Ken French’s, validating complex multivariate models like this would be impossible, leaving the field reliant on trust rather than verification.
