# 1. Original Study Summary

## (a) Research Question: Clearly state the paper’s main research question and economic Motivation

Modern asset-pricing began with a simple premise: if markets are competitive and information flows freely, prices should adjust to news and leave no predictable profit opportunities. This idea is what became the **Efficient Market Hypothesis (EMH)**. EMH emerged in the 1950s and 60s out of early empirical work showing that stock returns had almost no autocorrelation, suggesting prices behave like a fair game responding only to new information.

A decade later, Sharpe (1964) and Lintner (1965) formalized the first equilibrium model linking risk to expected returns: the **Capital Asset Pricing Model (CAPM)**. Its prediction was:

**(1) CAPM Formula**  
\[$
E[R_i] - R_f = \beta_i\,(E[R_M] - R_f)
$\]

With **beta** as the sole predictor of cross-sectional differences in average returns. For 20 years this model remained empirically reasonable.

But as Fama repeatedly emphasizes, CAPM’s “golden age” was short. By the early 1980s and into the 1990s, empirical papers accumulated showing that beta alone could not explain average returns sorted on firm **size** (Banz 1981), **earnings-price ratios** (Basu 1977), **leverage** (Bhandari 1988), or **book-to-market** (Rosenberg, Reif & Lanstein 1985). Additional patterns such as long-term reversals (DeBondt & Thaler 1985) and the January effect (Keim 1983) reinforced the conclusion that CAPM missed persistent dimensions of return variance.  

These anomalies eventually shifted professional consensus. As the *New York Times* summarized in 1992—reflecting the conclusions of Fama and French—“beta as the sole variable explaining returns on stocks is dead.” (Berg 1992)

These empirical failures exposed a deeper theoretical problem: CAPM offered too narrow a representation of risk. Asset-pricing models required additional dimensions to explain why some portfolios earned persistently higher returns than others.

When Eugene Fama and Kenneth French aimed at capturing these missing dimensions, they developed the **Fama–French Three-Factor Model (1993)**. In it they augmented the market factor with size (SMB: small minus big) and value (HML: high minus low book-to-market):

**(2) Fama–French Three-Factor Formula**  
\[$
R_i - R_f = \alpha_i + \beta_i (R_M - R_f) + s_i\,\text{SMB} + h_i\,\text{HML} + \varepsilon_i
$\]

What the Three-Factor Model aims to answer is: can size (SMB) and value (HML) jointly explain the cross-section of average returns better than CAPM?

---

After the Three-Factor Model (1993), Fama and French showed that SMB and HML captured important patterns that CAPM could not explain. But the model was still incomplete: it summarized the data, yet it did not explain **why** these patterns existed. Subsequent work showed that **profitability** and **investment** were also strongly related to average returns and were not absorbed by the three factors.

The motivation for revisiting the model comes directly from the **dividend discount model**, which expresses a stock’s price as the present value of expected future dividends:

**(3) Dividend Discount Model**  
\[$
P_t = \sum_{j=1}^{\infty} \frac{E_t[D_{t+j}]}{(1+r)^j}
$\]

The DDM can be rewritten in terms of future earnings and changes in book equity:

**(4) \(M_t\)**  
\[$
M_t = \frac{B_{t+1} - B_t + D_{t+1}}{E_t[R_{t+1}]}
$\]

This implies three core relationships:
- A higher book-to-market ratio is associated with **higher expected returns**  
- Higher expected profitability is associated with **higher expected returns**  
- Higher expected investment (growth in book equity) is associated with **lower expected returns**

This leads directly to the research question in the **Five-Factor Model of Fama and French (2014):**

> **“Do profitability and investment add explanatory power beyond the three-factor model?”**

The economic interpretation is straightforward: if expected returns are tied to how firms earn profits and reinvest capital, then an empirical model that omits these drivers is incomplete. The purpose of the five-factor model is to test whether incorporating **profitability (RMW)** and **investment (CMA)** produces a model better aligned with underlying expected returns.


## (b) Theoretical Framework: Summarize the economic theory underlying the analysis


The primary economic theory that the paper's analysis is based on is the dividend discount model. The authors state this directly: "We can use the dividend discount model to explain why these variables are related to average returns.” (Fama and French, 2014, p. 1)

This model states that the price of the stock is the discounted value of its expected future dividends. The authors then mathematically rearrange this model into a valuation equation (Equation 3) that expresses the ratio of a firm's market value to its book value (Mt /Bt ) as a function of its expected future earnings and expected future investments (changes in book equity) .

This valuation equation provides the theoretical justification for the new factors by generating three specific predictions about expected returns . As the paper states:

“Eq. (3) makes three statements about expected stock returns. First, fix everything in (3) except the current value of the stock, Mt , and the expected stock return, r. Then a lower value of Mt , or equivalently a higher book-to-market equity ratio, Bt /Mt , implies a higher expected return. Next, fix Mt  and the values of everything in (3) except expected future earnings and the expected stock return. The equation then tells us that higher expected earnings imply a higher expected return. Finally, for fixed values of Bt , Mt , and expected earnings, higher expected growth in book equity investment implies a lower expected return." (Fama and French, 2014, p. 2)

In short, the dividend discount model provides the theoretical basis for why value (high B/M), profitability (high earnings), and investment (low investment) should all be related to higher average stock returns. These three predictions are the "natural choices" that motivate adding profitability and investment factors to the model.	

## (c) Methodology: Describe the econometric approach, model specification, and estimation technique

The paper's core model specification is the five-factor asset pricing model. The authors explicitly define this in a regression equation:
"Motivated by this evidence and the valuation Eq. (3), we add profitability and investment factors to the three-factor model:
$R_{it}-R_{Ft}=a_{i}+b_{i}(R_{Mt}-R_{Ft})+s_{i}SMB_{t}+h_{i}HML_{t}+r_{i}RMW_{t}+c_{i}CMA_{t}+e_{it}$" (Fama and French, 2014, p. 3)

In this equation, the dependent variable ($R_{it}−R_{Ft}$) is the excess return on a portfolio or security i at time t. The independent variables are the five factors: the market excess return ($R_{Mt}−R_{Ft}$), the size factor ($SMBt$), the value factor ($HMLt$), the profitability factor ($RMWt$), and the investment factor ($CMAt$).
The estimation technique used is a time-series regression. This regression is run for each of the left-hand-side ($LHS$) portfolios (e.g., the 25 Size-B/M portfolios) against the five factor returns on the right-hand-side ($RHS$). The output of this estimation provides the factor loadings (the b, s, h, r, and c coefficients) for each portfolio, as well as the intercept, $a_i$.

The econometric approach is designed to test how well the model explains average returns. The central hypothesis of this approach is that the model's factors should capture all systematic variation in average returns, leaving no unexplained portion. The paper states this testable implication clearly:
"If the exposures to the five factors, $b_i$, $s_i$, $h_i$, $r_i$, and $c_i$, capture all variation in expected returns, the intercept $a$ in (5) is zero for all securities and portfolios i." (Fama and French, 2014, p. 3)

Therefore, the primary goal is to test whether these intercepts ($a_i$), also known as "alphas," are statistically indistinguishable from zero. To do this for a set of portfolios simultaneously, the authors employ a specific, formal statistical test, as noted here:
"Table 5 shows the GRS statistic of Gibbons, Ross, and Shanken (1989) that tests this hypothesis for combinations of LHS portfolios and factors." (Fama and French, 2014, p. 9)

The estimation technique used to obtain the intercepts and factor exposures is identified as a time-series regression (this quote refers to the three-factor model, but the same technique is applied to the five-factor model):
“Tests of the three-factor model center on the time-series regression:
$R_{it}-R_{ft}=a_i+b_i(R_{mt}-R_{ft})+s_iSMB_t+h_iHML_t+e_{it}$” (Fama and French, 2014, p. 3)


## (d) Key Variables: Define dependent and independent variables with their economic interpretation

The dependent variable in the regressions is the **excess return on a security or portfolio ($R_{it}−R_{Ft}$).**
In this equation $R_{it}$ is the return on security or portfolio $i$ for period $t$, $R_{FΩ}$ is the riskfree returnn.

The independent variables are the five factors used to explain the dependent variable's returns.


**Market Factor ($R_{Mt}−R_{Ft}$):**
"$R_{Mt}$ is the return on the value-weight (VW) market portfolio..." (subtracted by the risk-free return, $R_{Ft}$).

- **Economic Interpretation:** This captures the overall market risk.


**Size Factor ($SMB$):**
"$SMB_t$ is the return on a diversified portfolio of small stocks minus the return on a diversified portfolio of big stocks..."

- **Economic Interpretation:** This captures the "size premium," or the difference in returns between small-cap and large-cap firms.


**Value Factor ($HML$):**
"$HML_t$ is the difference between the returns on diversified portfolios of high and low B/M stocks..."

- **Economic Interpretation:** This captures the "value premium," or the difference in returns between firms with high book-to-market ratios (value stocks) and low book-to-market ratios (growth stocks).


**Profitability Factor ($RMW$):**
"$RMW_t$ is the difference between the returns on diversified portfolios of stocks with robust and weak profitability..."

- **Economic Interpretation:** This captures the "profitability premium," or the difference in returns between firms with high (robust) profitability and low (weak) profitability.


**Investment Factor ($CMA$):**
"$CMA_t$ is the difference between the returns on diversified portfolios of the stocks of low and high investment firms, which we call conservative and aggressive."

- **Economic Interpretation:** This captures the "investment premium," or the difference in returns between firms that invest conservatively (low investment) and those that invest aggressively (high investment).


## (e) Main Findings: Summarize the authors’ primary empirical results and conclusions

The authors' primary empirical finding is that their new model is a clear improvement over their previous one. As they state, "A five-factor model directed at capturing the size, value, profitability, and investment patterns in average stock returns performs better than the three-factor model of Fama and French (FF, 1993)".


However, this improved performance does not mean the model is perfect. While it is a better description of returns, it is still statistically rejected as a complete model. The authors write, "The GRS test easily rejects a five-factor model directed at capturing these patterns, but we estimate that the model explains between 71% and 94% of the cross-section variance of expected returns for the Size, B/M, OP, and Inv portfolios we examine". (Fama and French, 2014, p. 9)


One of the most striking conclusions is that the addition of the new profitability and investment factors makes the original value factor (HML) unnecessary for explaining average returns.
- "With the addition of profitability and investment factors, the value factor of the FF three-factor model becomes redundant for describing average returns in the sample we examine." (Fama and French, 2014, p. 1)


The reason for this redundancy is that the information captured by the value factor is also captured by the other factors in the model.
- "The average HML return is captured by the exposures of HML to other factors". (Fama and French, 2014, p. 2) The authors elaborate on this, concluding, "...our results suggest that HML is a redundant factor in the sense that its high average return is fully captured by its exposures to $R_M−R_F$, SMB, and especially RMW and CMA." (Fama and French, 2014, p. 19)


Despite its successes, the authors clearly identify the model's main failure point, which they term its "lethal" problem.
- "The five-factor model's main problem is its failure to capture the low average returns on small stocks whose returns behave like those of firms that invest a lot despite low profitability."(Fama and French, 2014, p. 1)


This specific type of stock portfolio consistently produces results that the model cannot explain.
- "One of our more interesting results is that portfolios of small stocks with negative exposures to RMW and CMA are the biggest asset pricing problem in four of the six sets of LHS portfolios examined here."(Fama and French, 2014, p. 19)


From this, the authors draw a firm conclusion about the nature of this "lethal" portfolio: 
- "...we conclude that the lethal portfolios contain small stocks of firms that invest a lot despite low profitability". (Fama and French, 2014, p. 19)

### (d) Variable Construction:

* **Independent Variables (RHS Factors):**
    * **$R_M-R_F$**: The market excess return when compared to risk free asset return?


    * **SMB (Small Minus Big):** The size factor.

        * What it represents: This factor captures the size premium, which is the historical observation that small-company stocks (small market capitalization) tend to have higher average returns than large-company stocks.
     
            * How it's built:
              * Firms are sorted based on their market capitalization (Size).
              * The factor buys (goes long) a diversified portfolio of small stocks.
              * It sells (goes short) a diversified portfolio of big stocks.
                
          * What it means:
              * A positive SMB return means small-cap stocks outperformed large-cap stocks for that period.
              * A portfolio with a positive SMB beta is tilted toward small-cap stocks.


    * **HML (High Minus Low):** The value factor.

        * What it represents: This factor captures the value premium. This is the historical observation that "value" stocks (which appear cheap) tend to outperform "growth" stocks (which appear expensive).
          
         * How it's built:
             * Firms are sorted based on their Book-to-Market (B/M) ratio (the accounting value of the company divided by its stock market value).
             * A High B/M ratio means the stock is "cheap" (a value stock).
             * A Low B/M ratio means the stock is "expensive" (a growth stock).
             * The factor buys (goes long) a portfolio of High B/M (value) stocks.
             * It sells (goes short) a portfolio of Low B/M (growth) stocks.
           
        * What it means:
            * A positive HML return means value stocks outperformed growth stocks for that period.
            * A portfolio with a positive HML beta is tilted toward value stocks. A negative beta means it's tilted toward growth stocks.


    * **RMW (Robust Minus Weak):** The profitability factor.
 
        * What it represents: This factor captures the profitability premium. This is the observation that firms with high (robust) profitability tend to have higher average returns than firms with low (weak) profitability.
          
        * How it's built:
     
            * Firms are sorted based on their Operating Profitability (OP), which the paper measures as revenues minus costs, all divided by book equity.
            * The factor buys (goes long) a portfolio of Robust Profitability stocks.
            * It sells (goes short) a portfolio of Weak Profitability stocks.
              
        * What it means:
            * A positive RMW return means high-profitability firms outperformed low-profitability firms.
            * A portfolio with a positive RMW beta is tilted toward "quality" or high-profitability firms.

     
    * **CMA (Conservative Minus Aggressive):** The investment factor.

        * What it represents: This factor captures the investment premium. This is based on the theory and observation that companies that invest conservatively (grow their assets slowly) tend to have higher returns than firms that invest aggressively (grow their assets very quickly).
 
        * How it's built:
            * Firms are sorted based on their investment rate, which the paper measures as the year-over-year growth in total assets.
            * The factor buys (goes long) a portfolio of low-investment (Conservative) firms.
            * It sells (goes short) a portfolio of high-investment (Aggressive) firms.
              
        * What it means:
            * A positive CMA return means low-investment firms outperformed high-investment firms.
            * A portfolio with a positive CMA beta is tilted toward firms that are more conservative with their asset growth.

## 3. Replication Analysis

### (a) Model Specification: 

## (a) Model Specification

We estimate the Fama–French Five-Factor Model, which explains the excess return of a portfolio using five systematic risk factors. The regression equation is:

$$
R_{it} - R_{ft}
= \alpha_i
+ b_i \, (R_{Mt} - R_{ft})
+ s_i \, SMB_t
+ h_i \, HML_t
+ r_i \, RMW_t
+ c_i \, CMA_t
+ \varepsilon_{it}.
$$


where:

- **$R_{it} - R_{ft}$**  
  Portfolio *i*’s **excess return** at time *t*.  
  This is the dependent variable of the regression.

- **$\alpha$**  
  The portfolio’s **abnormal return** (intercept).  
  If the model prices the portfolio correctly, $\alpha_i$ should be statistically indistinguishable from zero.

- **$R_{Mt} - R_{ft}$** (MKT-RF)  
  The **market excess return**.  
  Captures the portfolio’s exposure to overall market risk.

- **$SMB_t$**  
  “Small Minus Big.”  
  The **size factor**, measuring the excess return of small firms over large firms.

- **$HML_t$**  
  “High Minus Low.”  
  The **value factor**, measuring the return spread between high book-to-market (value) and low book-to-market (growth) firms.

- **$RMW_t$**  
  “Robust Minus Weak.”  
  The **profitability factor**, capturing the return difference between firms with robust vs. weak operating profitability.

- **$CMA_t$**  
  “Conservative Minus Aggressive.”  
  The **investment factor**, measuring the return spread between firms that invest conservatively and those that invest aggressively.

- **$\varepsilon_{it}$**  
  Regression error term.

 **Coefficients**
- $b_i$: loading on the market factor  
- $s_i$: loading on the size factor  
- $h_i$: loading on the value factor  
- $r_i$: loading on the profitability factor  
- $c_i$: loading on the investment factor
  

These coefficients measure how sensitively the portfolio’s excess return responds to movements in each factor. The model is estimated using monthly data and Newey–West (HAC) standard errors to correct for potential autocorrelation and heteroskedasticity.


**Loading the Model:**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats

factors = pd.read_csv('5f_data.csv', skiprows=4)
factors = factors[pd.to_numeric(factors['Unnamed: 0'], errors='coerce').notna()]
factors = factors.rename(columns={'Unnamed: 0': 'date', 'Mkt-RF': 'MKTRF'})
factors = factors[['date', 'MKTRF', 'SMB', 'HML', 'RMW', 'CMA', 'RF']]
factors.iloc[:, 1:] = factors.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

port = pd.read_csv('25_portfolios_5x5.csv', skiprows=15)
port = port.rename(columns={'Unnamed: 0': 'date'})
port = port[pd.to_numeric(port['date'], errors='coerce').notna()]
port = port[:-1]
port.iloc[:, 1:] = port.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

factors = factors.drop_duplicates(subset='date')
port = port.drop_duplicates(subset='date')

data = pd.merge(port, factors, on='date', how='inner')
````

### (b) Estimation Results:

**Regression of Small, Low Book-to-Market Portfolios**

````python
cols = ['SMALL LoBM', 'RF', 'MKTRF', 'SMB', 'HML', 'RMW', 'CMA']
data[cols] = data[cols].apply(pd.to_numeric, errors='coerce')

data['excess'] = data['SMALL LoBM'] - data['RF']
data['excess'] = pd.to_numeric(data['excess'], errors='coerce')

reg = smf.ols('excess ~ MKTRF + SMB + HML + RMW + CMA',
              data=data).fit(cov_type='HAC', cov_kwds={'maxlags':12})

reg.summary()
````

````python
factor_cols = ['MKTRF', 'SMB', 'HML', 'RMW', 'CMA', 'RF']
portfolio_cols = [c for c in data.columns if c not in factor_cols + ['date']]

results = []

for port in portfolio_cols:

    excess = pd.to_numeric(data[port], errors='coerce') - pd.to_numeric(data['RF'], errors='coerce')
    data['excess'] = excess


    reg = smf.ols('excess ~ MKTRF + SMB + HML + RMW + CMA',
                  data=data).fit(cov_type='HAC', cov_kwds={'maxlags':12})

    results.append({
        'Portfolio': port,
        'Alpha': reg.params['Intercept'],
        't(Alpha)': reg.tvalues['Intercept'],
        'b_MKT': reg.params['MKTRF'],
        't(MKT)': reg.tvalues['MKTRF'],
        'b_SMB': reg.params['SMB'],
        't(SMB)': reg.tvalues['SMB'],
        'b_HML': reg.params['HML'],
        't(HML)': reg.tvalues['HML'],
        'b_RMW': reg.params['RMW'],
        't(RMW)': reg.tvalues['RMW'],
        'b_CMA': reg.params['CMA'],
        't(CMA)': reg.tvalues['CMA'],
        'R2': reg.rsquared
    })


results_df = pd.DataFrame(results)
results_df = results_df.set_index('Portfolio')
results_df
````
