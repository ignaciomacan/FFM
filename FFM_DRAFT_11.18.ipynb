<h1 style="font-family: Garamond; font-weight: bold; font-size: 28px; margin-bottom: 0.4em;">
1. Original Study Summary
</h1>

<h2 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
(a) Research Question: Clearly state the paper’s main research question and economic motivation
</h2>


Modern asset-pricing began with a simple premise: if markets are competitive and information flows freely, prices should adjust to news and leave no predictable profit opportunities. This idea is what became the **Efficient Market Hypothesis (EMH)**. EMH emerged in the 1950s and 60s out of early empirical work showing that stock returns had almost no autocorrelation, suggesting prices behave like a fair game responding only to new information.

A decade later, Sharpe (1964) and Lintner (1965) formalized the first equilibrium model linking risk to expected returns: the **Capital Asset Pricing Model (CAPM)**. Its prediction was:

<h4 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.2em;">
(1) CAPM Formula
</h4>


$$
E[R_i] - R_f = \beta_i\,(E[R_M] - R_f)
$$

CAPM originally told us that one thing explains expected returns: **market risk**. 

With **beta** as the sole predictor of cross-sectional differences in average returns. For 20 years this model remained empirically reasonable.

---

But as Fama repeatedly emphasizes, CAPM’s “golden age” was short. By the early 1980s and into the 1990s, empirical papers accumulated showing that the relationship between market betas and average returns was basically "flat", but according to CAPM it couldn't be flat, average returns should increase with market beta. So beta alone could not explain average returns sorted on firm **size** (Banz 1981), **earnings-price ratios** (Basu 1977), **leverage** (Bhandari 1988), or **book-to-market** (Rosenberg, Reif & Lanstein 1985).

These accumulating anomalies eventually shifted professional consensus. In a 1992 New York Times article based on a phone interview with Eugene Fama, he stated that “beta as the sole variable explaining returns on stocks is dead” (Berg 1992).

These empirical failures exposed a deeper theoretical problem: CAPM offered too narrow a representation of risk. Asset pricing models required additional dimensions to explain why some portfolios earned persistently higher returns than others.

When Eugene Fama and Kenneth French aimed at capturing these missing dimensions, they developed the **Fama–French Three-Factor Model (1993)**, which looked at the two biggest anomalies for the CAPM, which was it's inability to explain small stock returns and the difference between value and growth stock returns. That's why in the Three Factor Model they augmented the market factor with size (SMB: small minus big) and value (HML: high minus low book-to-market):
  
<h4 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.2em;">
(2) Fama-French Three Factor Formula
</h4>

$$
R_i - R_f = \alpha_i + \beta_i (R_M - R_f) + s_i\,\text{SMB} + h_i\,\text{HML} + \varepsilon_i
$$

<h4 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.2em;">
Research Question of the Three-Factor Model
</h4>

The central question motivating the Three-Factor Model is whether **size (SMB)** and **value (HML)** contain systematic information about expected returns that the CAPM’s single beta cannot capture. In other words:

**Can the cross section of average stock returns be better explained by adding size and value factors to the market factor, thereby accounting for the persistent return premiums observed for small firms and high book to market firms?**

After the Three-Factor Model (1993), Fama and French showed that the size and value factors **SMB** and **HML** captured important patterns that the CAPM could not explain. Small firms earned higher average returns than big firms (**size premium**), and high book to market firms earned higher returns than growth firms (**value premium**). In the three-factor framework, these effects appear through the parameters $( s_i )$ and $( h_i )$:

 $( s_i > 0 )$ for small firms and $( s_i < 0 )$ for big firms  
 $( h_i > 0 )$ for high book-to-market (value) firms

But the model was still incomplete as evident by Novy-Marx (2012), Titman, Wei and Xie (2004) which say that the Three Factor is incomplete model for expected returns because its three factors miss much of the variation in average returns related to profitability and investment. 

---
A natural next step after the Three-Factor Model was to ask whether basic valuation logic could justify adding additional predictors. Under the standard dividend-discount framework,

$
m_t = \sum_{k=1}^{\infty} \frac{E_t[d_{t+k}]}{(1+r)^k}
$

**holding other characteristics constant**, a firm with higher expected profitability should have a higher required return $( r )$. Put differently, if two firms have similar book equity $( B )$ but one has higher expected earnings or cash flows, the valuation identity implies a positive link between profitability and expected returns. Empirically, portfolios sorted on operating profitability $(\text{OP})$ showed exactly this pattern.

Likewise, investment behavior carries information about expected returns. In standard steady state valuation, higher investment $( I_t )$ relative to assets $( A_t )$ (i.e., high asset growth) is typically associated with lower expected returns because firms invest more when their marginal $( q )$ is low. This motivates a negative relation between investment and expected returns. Fama and French operationalize this using the investment factor:

$
\text{CMA} = R_{\text{Conservative}} - R_{\text{Aggressive}}
$

Combining theis economic motivation, Fama and French extend the Three-Factor Model by adding profitability and investment factors:

<h4 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.2em;">
(3) Fama-French Five Factor Formula
</h4>

$$
R_{i,t} - R_{f,t}
= \alpha_i
+ \beta_i \left(R_{M,t} - R_{f,t}\right)
+ s_i\,\text{SMB}_t
+ h_i\,\text{HML}_t
+ r_i\,\text{RMW}_t
+ c_i\,\text{CMA}_t
+ \varepsilon_{i,t}.
+ $$
<h4 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.2em;">
Research Question of the Five-Factor Model
</h4>

The Five-Factor Model asks whether **profitability (RMW)** and **investment (CMA)**, two characteristics central to standard valuation theory, further improve our understanding of the cross-section of expected returns beyond size and value. Meaning:

**Do differences in firm's profitability and investment behavior help explain average returns in a way that is consistent with basic discounted cash flow logic, and do these two factors capture return patterns not absorbed by the original three factors?**

The Five-Factor Model tests whether expected returns vary **positively with profitability** and **negatively with investment**, completing the move from anomaly driven factors toward factors grounded in economic fundamentals.

In this Five Factor Model, the economic prediction and what the data confirm is that **average returns vary positively with profitability** $(r_i > 0)$ **and negatively with investment** $(c_i < 0)$.

<h2 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
(b) Theoretical Framework: Summarize the economic theory underlying the analysis
</h2>

<h2 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
0. CAPM
</h2>

The Capital Asset Pricing Model (CAPM) predicts that expected excess returns depend only on market risk:

$E[R_i - R_f] = \beta_i \, E[R_M - R_f].$

Empirically, we estimate:

$R_{i,t} - R_{f,t}
= \alpha_i + \beta_i (R_{M,t} - R_{f,t}) + \varepsilon_{i,t}.$


**CAPM said that market beta *$\beta$* should fully explain expected returns**, leaving no room for abnormal performance *$\alpha$*.  
In theory, **$\alpha$** is the extra return beyond what your risk exposure predicts, the “bonus” return, but it shouldn’t exist if the model is good at explaining returns.

However, when researchers tested CAPM, they found return patterns, especially for small firms and high book-to-market firms that produced large, persistent **abnormal returns** (*$\alpha$*).  
These anomalies revealed that CAPM was missing important sources of risk, motivating the move to the **Three-Factor Model**.


---

<h2 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
1. Fama–French Three-Factor Model
</h2>

<p style="text-align: center; font-family: Garamond; font-weight: bold; font-size: 20px;">
3 Factor Regression
</p>

$$
R_{i,t} - R_{f,t}
= \alpha_i
+ b_i (R_{M,t} - R_{f,t})
+ s_i \, \text{SMB}_t
+ h_i \, \text{HML}_t
+ \varepsilon_{i,t}.
$$

**Outcome variable:**  
$(R_{i,t} - R_{f,t})$: excess return on portfolio $i$ at time $t$.

**X Variables:**

**Market $(R_M - R_f)$**  
Represents aggregate systematic risk (the original CAPM beta risk).

**SMB (Size)**  
Small firms tend to have more volatile and cyclical earnings, weaker balance sheets, and greater sensitivity to recessions.  
Higher exposure to SMB reflects greater “small-firm risk,” so portfolios with larger SMB loadings should earn higher expected returns:

$s_i \uparrow \;\Rightarrow\; E[R_i] \uparrow$

**HML (Value)**  
High Book-to-Market (value) firms typically have low prices relative to fundamentals, persistently weak earnings, and characteristics associated with financial distress.  
Low Book-to-Market (growth) firms tend to be more profitable and less distressed.

If markets price risk rationally, portfolios with higher HML loadings are more exposed to “value/distress” risk and therefore earn higher expected returns:

$h_i \uparrow \;\Rightarrow\; E[R_i] \uparrow$

The three-factor model is thus a **risk-based extension of CAPM** that absorbs the size and value patterns left unexplained by beta alone.

---

<h2 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
2. Fama–French Five-Factor Model
</h2>

<p style="text-align: center; font-family: Garamond; font-weight: bold; font-size: 20px;">
5 Factor Regression
</p>

$$
R_{i,t} - R_{f,t}
= \alpha_i
+ b_i (R_{M,t} - R_{f,t})
+ s_i \, \text{SMB}_t
+ h_i \, \text{HML}_t
+ r_i \, \text{RMW}_t
+ c_i \, \text{CMA}_t
+ \varepsilon_{i,t}.
$$

**Outcome variable:**  
$(R_{i,t} - R_{f,t})$: excess return on portfolio $i$ at time $t$.

**X Variables:**

Market $(R_M - R_f)$ – aggregate systematic risk.  

SMB – small-minus-big size factor.  

HML – high-minus-low value factor.  

**RMW – Robust minus Weak profitability.**  

**CMA – Conservative minus Aggressive investment.**

---

<h3 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
Valuation Economic Interpretation
</h3>

The Five-Factor Model is built directly from a simple valuation identity: a stock’s price must equal the discounted value of its future cashflows. In the dividend discount model,

$$
m_t = \sum_{\tau=1}^{\infty} \frac{E(d_{t+\tau})}{(1+r)^\tau},
$$

the price *must* reflect expectations of future dividends and the discount rate $r$, the stock’s long-run expected return. If two firms are expected to pay the same future dividends but one is cheaper today, the cheaper firm must have a higher expected return. This is the basic insight behind why valuation ratios like B/M forecast returns.

Miller and Modigliani (1961) show that the same logic applies to total equity, where cashflows can be decomposed into earnings minus investment:

$$
M_t = \sum_{\tau=1}^{\infty} 
\frac{E(Y_{t+\tau} - dB_{t+\tau})}{(1+r)^\tau},
$$

$$
\frac{M_t}{B_t}
=
\sum_{\tau=1}^{\infty}
\frac{E(Y_{t+\tau} - dB_{t+\tau})}{B_t (1+r)^\tau}.
$$

The third equation tells us that expected returns depend on three ingredients:

**(1) Valuation:** higher $B_t/M_t$ (cheap stocks) imply higher expected returns.  

**(2) Profitability:** higher expected earnings $E(Y_{t+\tau})$ imply a higher expected return.  

**(3) Investment:** higher expected growth in book equity $dB_{t+\tau}$ implies a lower expected return, because investing heavily today raises future earnings but lowers near-term cashflows available to shareholders.

The economic challenge, historically, was finding real-world proxies for expected profitability and expected investment. Novy-Marx (2013) shows that operating profitability strongly predicts average returns, and Aharoni, Grundy, and Zeng (2013) show that investment does as well. These patterns are only weakly captured by the original three-factor model.

Fama and French’s key insight is that RMW and CMA are the *natural* empirical proxies implied by the valuation decomposition: RMW captures differences in expected profitability, and CMA captures differences in expected investment. Together, they extend the three-factor model so that it lines up with the fundamental accounting identity linking prices, earnings, and investment.

---

**Profitability (RMW)**  

Higher expected earnings raise the numerator of the valuation identity.  
To keep the observed price fixed, the discount rate (expected return) must rise:

$E(Y_{t+\tau}) \uparrow \;\Rightarrow\; r \uparrow$

Portfolios with higher RMW loadings are more exposed to profitability risk and should earn higher expected returns:

$r_i \uparrow \;\Rightarrow\; E[R_i] \uparrow$

**Investment (CMA)**  

Higher investment increases future book equity, reducing net cashflow $(Y_{t+\tau} - dB_{t+\tau})$ for a given price.  
This pushes the required return **down**:

$dB_{t+\tau} \uparrow \;\Rightarrow\; r \downarrow$

Portfolios with higher CMA loadings behave like firms that invest conservatively (low $dB$) and thus carry higher expected returns than aggressively investing firms:

$c_i \uparrow \;\Rightarrow\; E[R_i] \downarrow$


<h2 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
(c) Methodology: Econometric Approach, Model Specification, and Estimation
</h2>

The econometric approach follows the standard Fama–French asset pricing methodology. On the right-hand side (RHS), each model uses a set of traded factor returns, market excess return for CAPM. Market, SMB, and HML for the Three-Factor Model; and market, SMB, HML, RMW, and CMA for the Five-Factor Model to explain variation in portfolio excess returns. We estimate factor loadings (aka the time-series regression slopes) using time-series OLS regressions of portfolio excess returns on factor returns, and we evaluate model performance through the main diagnostics used in Fama and French (1993, 2015): the magnitude and significance of intercepts, the signs and t-statistics of factor loadings, the regression $R^2$ values that indicate how much common variation the factors capture, and the joint GRS test of whether all alphas are zero. We also compare models using the variance of the estimated alphas $alpha_i$, and examine whether adding profitability and investment factors reduces the magnitude and cross-sectional variance of these pricing errors across the 25 portfolio and 32 portfolio sorts.

We estimate factor loadings using time-series Ordinary Least Squares (OLS) regressions of portfolio excess returns on factor returns. For each test portfolio $i$, we run a separate regression of monthly excess returns $(R_{it} - R_{Ft})$ on the factor returns. This produces the intercept $(a_i)$ and the slope coefficients $(b_i, s_i, h_i, r_i, c_i)$, which measure the portfolio’s exposure to the market, size, value, profitability, and investment factors.

Our left side (LHS) portfolios follow the Fama and French (2015) design. Portfolios are formed from independent sorts on Size, Book-to-Market (B/M), Operating Profitability (OP), and Investment (Inv), and are then used as LHS assets in the regressions:

- 25 Size–B/M portfolios (5 x 5 sort on Size and B/M)  
- 25 Size–OP portfolios (5 x 5 sort on Size and operating profitability)  
- 25 Size–Inv portfolios (5 x 5 sort on Size and investment)  
- 32 Size–B/M–OP portfolios (2 x 4 x 4 sort on Size, B/M, and OP)  
- 32 Size–B/M–Inv portfolios (2 x 4 x 4 sort on Size, B/M, and investment)  
- 32 Size–OP–Inv portfolios (2 x 4 x 4 sort on Size, OP, and investment)

For each of these six 'left side' sets, we estimate the same sequence of time-series regressions on the right side (x) factors:

- CAPM: market excess return only  
- Three-Factor Model: market, SMB, HML  
- Five-Factor Model: market, SMB, HML, RMW, CMA  

This means, for example, that the 25 Size–B/M portfolios generate 25 separate CAPM regressions, 25 three-factor regressions, and 25 five-factor regressions, and similarly for each of the other 25 portfolio and 32 portfolio sorts. In total, this yields hundreds of time-series regressions, as in Fama and French (2015).

In the original three-factor paper (Fama and French, 1993), the same time-series OLS framework is applied to 25 Size–B/M stock portfolios and 7 bond portfolios. In our replication, we focus on the equity portfolios and estimate the three-factor model and the five-factor model on the same sets of stock portfolios, so model comparisons are based on identical LHS assets.

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Model Specification
</h3>

The Three-Factor Model is estimated using:

$$
R_{it} - R_{Ft}
= a_i
+ b_i (R_{Mt} - R_{Ft})
+ s_i \, SMB_t
+ h_i \, HML_t
+ e_{it}.
$$

For the Three-Factor Model, we follow the original Fama and French (1993) design and estimate the model **only** on the 25 Size–B/M portfolios. These portfolios are constructed from a 5 x 5 independent sort on Size and Book-to-Market and were the primary equity test assets used in the original three-factor paper. 

Thus, the Three-Factor Model produces **25 separate time-series regressions**, one for each Size–B/M portfolio.

<br>

The Five-Factor Model extends the Three-Factor Model by including profitability and investment:

$$
R_{it} - R_{Ft}
= a_i
+ b_i (R_{Mt} - R_{Ft})
+ s_i \, SMB_t
+ h_i \, HML_t
+ r_i \, RMW_t
+ c_i \, CMA_t
+ e_{it}.
$$

Once we move to the Five-Factor Model, we expand the left-hand-side (LHS) assets to match the full testing design of Fama and French (2015). Profitability and investment sorts did not exist in the earlier three-factor model, so they enter the analysis only at this stage.

The Five-Factor Model is estimated on the six portfolio families introduced in Fama and French (2015):

- 25 Size–B/M portfolios  
- 25 Size–OP portfolios  
- 25 Size–Inv portfolios  
- 32 Size–B/M–OP portfolios  
- 32 Size–B/M–Inv portfolios  
- 32 Size–OP–Inv portfolios  

Each portfolio in each sort generates one time-series regression. This produces **171 separate regressions** per model specification and allows direct comparison of the Three-Factor and Five-Factor Models on a greatly expanded set of characteristics.

This structure mirrors the original papers: the Three-Factor Model is evaluated on the traditional 25 Size–B/M portfolios, while the Five-Factor Model is evaluated on the complete set of 25-portfolio and 32-portfolio sorts that incorporate profitability and investment information.


The coefficients $(b_i, s_i, h_i, r_i, c_i)$ represent exposures to the underlying economic risks, while $(a_i)$ measures the model’s pricing error for portfolio $i$.

Each coefficient captures a different dimension of risk:

- **$b_i$ (Market):** sensitivity to the excess market return $(R_M - R_f)$.
- **$s_i$ (Size / SMB):** exposure to the return spread between small and big firms.
- **$h_i$ (Value / HML):** exposure to the return spread between high and low book-to-market firms.
- **$r_i$ (Profitability / RMW):** sensitivity to the return spread between robust and weak profitability firms.
- **$c_i$ (Investment / CMA):** sensitivity to the return spread between conservative and aggressive investment firms.

Together, these loadings summarize how each portfolio co-moves with the factor premiums emphasized by Fama and French.


<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Sample
</h3>

We use **monthly portfolio returns** and **monthly factor returns**, consistent with the original papers. All portfolio returns are converted to excess returns using the one-month Treasury bill rate. Our sample period matches the availability of the Fama–French Five-Factor dataset, and all portfolios are value-weighted as the author states in the data library.

The 25 portfolio sorts come from independent 5 x 5 sorts on Size and B/M, Size and OP, and Size and Inv. The 32 portfolio sorts come from 2 x 4 x 4 sorts on Size, B/M, and OP; Size, B/M, and Inv; and Size, OP, and Inv. These are the same portfolio definitions used in Fama and French (2015) to evaluate the three-factor and five-factor models.



<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Estimation Technique
</h3>

To test whether all alphas are jointly zero within a given portfolio set (for example, all 25 Size–B/M portfolios), we use the **GRS test** of Gibbons, Ross, and Shanken (1989), following Fama and French (2014). The GRS statistic is computed for each model and each of the six portfolio families.

Regression parameters are estimated using Ordinary Least Squares (OLS), consistent with Fama and French (1993, 2015). Statistical inference centers on the significance of the factor coefficients and the intercepts (alphas), where a correctly specified model implies $a_i = 0$ for all test portfolios. We report conventional OLS $t$-statistics, matching the approach in the original studies.

We also evaluate:

- whether the estimated alphas $(a_i)$ differ significantly from zero,  
- the explanatory power of each regression (adjusted $R^2$),  
- whether the signs and magnitudes of the factor loadings align with economic intuition,  
- and the overall fit of the model across different portfolio sets.

<h2 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
(d) Key Variables: Dependent and Independent Variables
</h2>

The dependent variable in all regressions is the **excess return on portfolio $i$ in month $t$**, defined as:

$$
Y_{it} = R_{it} - R_{Ft}.
$$

Here, $R_{it}$ is the value-weighted return on portfolio $i$ for month $t$, and $R_{Ft}$ is the one-month Treasury bill rate. Consistent with Fama and French (1993, 2015), all left hand side (LHS) portfolios are formed each June using NYSE breakpoints and their returns are computed monthly from July of year $t$ to June of year $t+1$. Value-weighted returns are computed as:

$$
R_{it} = \sum_{j \in \mathcal{P}_i} w_{j,t-1} R_{jt},
\qquad 
\sum_{j \in \mathcal{P}_i} w_{j,t-1} = 1.
$$

Our Y-portfolios follow the full Fama and French (2015) design. The test assets include:

- 25 Size–B/M portfolios  
- 25 Size–OP portfolios  
- 25 Size–Inv portfolios  
- 32 Size–B/M–OP portfolios  
- 32 Size–B/M–Inv portfolios  
- 32 Size–OP–Inv portfolios  

These portfolios produce large monotonic spreads across Size, Book to Market, Operating Profitability, and Investment. The key characteristics the models aim to explain.

<br>

<h3 style="font-family: Garamond; font-weight: bold; font-size: 18px; margin-bottom: 0.3em;">
Independent Variables (X): Factor Returns and Economic Interpretation
</h3>

The right hand side variables are the Fama–French factor returns. Each factor is a zero cost long–short portfolio intended to proxy for a systematic source of risk in expected returns. For each portfolio $i$, the estimated coefficients $(b_i, s_i, h_i, r_i, c_i)$ measure exposure to each factor.

<br>

**Market Factor (MKT):**  
$$
X_{1t} = R_{Mt} - R_{Ft}.
$$  
Economic interpretation: $b_i$ is the portfolio’s market beta, its sensitivity to aggregate economic risk.

<br>

**Size Factor (SMB):**  
$$
X_{2t} = SMB_t = R_{\text{Small},t} - R_{\text{Big},t}.
$$  
Economic interpretation: $s_i$ measures whether portfolio $i$ behaves like small firms ($s_i > 0$) or big firms ($s_i < 0$), reflecting the size premium.



<br>

**Value Factor (HML):**  
$$
X_{3t} = HML_t = R_{\text{High B/M},t} - R_{\text{Low B/M},t}.
$$  

HML measures the return premium investors earn for holding **value firms** rather than **growth firms**.

A **positive** $h_i$ means the portfolio behaves like value stocks and earns compensation for value-related risk.

A **negative** means the portfolio behaves more like growth stocks, which have lower expected returns.


<br>

**Profitability Factor (RMW):**  
$$
X_{4t} = RMW_t = R_{\text{Robust OP},t} - R_{\text{Weak OP},t}.
$$  

RMW measures the return premium associated with **profitability**, specifically the difference between firms with **robust operating profitability** and those with **weak profitability**.

A **positive** $r_i$ means the portfolio behaves like highly profitable firms and earns compensation for profitability related risk.

A **negative** means the portfolio behaves like weak profitability firms, which have lower and more volatile cash flows and therefore lower expected returns.


<br>

**Investment Factor (CMA):**  
$$
X_{5t} = CMA_t = R_{\text{Conservative Inv},t} - R_{\text{Aggressive Inv},t}.
$$  
Economic interpretation: $c_i$ reflects exposure to the investment premium. Conservatively investing firms earn higher expected returns than aggressively investing firms unless the latter have strong profitability.

<br>

Together, these factor loadings summarize how each portfolio co-moves with the underlying economic risks emphasized by Fama and French. A correctly specified factor model should absorb these systematic risks and leave intercepts that satisfy:

$$
a_i = 0 \quad \forall i.
$$


<h2 style="font-family: Garamond; font-weight: bold; font-size: 20px; margin-bottom: 0.3em;">
(e) Main Findings: Authors’ Primary Empirical Results and Conclusions
</h2>

The central empirical finding of Fama and French (2015) is that the Five-Factor Model offers a substantial improvement over the original Three-Factor Model. As they write, *“A five-factor model directed at capturing the size, value, profitability, and investment patterns in average stock returns performs better than the three-factor model of Fama and French (1993).”* The addition of profitability (RMW) and investment (CMA) factors sharply increases explanatory power and reduces the magnitude of pricing errors across most portfolio sets.

Despite this improvement, the model is not a perfect description of the data. The GRS test rejects the model as a complete asset-pricing model, even though its economic fit is strong. The authors note, *“The GRS test easily rejects a five-factor model directed at capturing these patterns, but we estimate that the model explains between 71% and 94% of the cross-section variance of expected returns for the Size, B/M, OP, and Inv portfolios we examine.”* In other words, the model is statistically rejected but economically powerful.

One of the most notable findings is that the profitability and investment factors render the original value factor (HML) redundant. The authors state directly that, *“With the addition of profitability and investment factors, the value factor of the FF three-factor model becomes redundant for describing average returns in the sample we examine.”* This redundancy arises because HML’s high average return can be almost entirely explained by its exposures to the other factors:  
*“…our results suggest that HML is a redundant factor in the sense that its high average return is fully captured by its exposures to $R_M - R_F$, SMB, and especially RMW and CMA.”*

This result highlights the theoretical success of the new factors:  
- **Profitability (RMW)** is strongly and positively related to average returns.  
- **Investment (CMA)** is strongly negatively related to average returns (firms investing aggressively earn lower returns).  

Together, these two factors subsume the information HML was capturing about cash-flow risk and investment behavior.

However, the Five-Factor Model also has a clear and well-defined failure. Fama and French describe this as the model’s *“lethal”* problem: its inability to explain the very low average returns on certain small, distressed firms. These are portfolios made up of small firms that **invest heavily despite having low profitability**, and they consistently generate pricing errors that the model cannot account for. As the authors write, *“Portfolios of small stocks with negative exposures to RMW and CMA are the biggest asset pricing problem in four of the six sets of LHS portfolios examined here.”* They conclude, *“…we find that the lethal portfolios contain small stocks of firms that invest a lot despite low profitability.”*

In summary, the main empirical outcomes are:

- The Five-Factor Model substantially improves upon the Three-Factor Model.  
- Profitability (RMW) is a powerful, economically meaningful factor with a strong positive relation to returns.  
- Investment (CMA) is also important, though it introduces challenges in capturing low-return, high-investment small firms.  
- The value factor (HML) becomes redundant once RMW and CMA are introduced.  
- Despite its successes, the model is still rejected statistically and fails most clearly for small, unprofitable, high-investment firms.

These findings form the basis of the paper’s contribution: a more complete but still imperfect model of the cross-section of expected returns.

### (d) Variable Construction:

* **Independent Variables (RHS Factors):**
    * **$R_M-R_F$**: The market excess return when compared to risk free asset return?


    * **SMB (Small Minus Big):** The size factor.

        * What it represents: This factor captures the size premium, which is the historical observation that small-company stocks (small market capitalization) tend to have higher average returns than large-company stocks.
     
            * How it's built:
              * Firms are sorted based on their market capitalization (Size).
              * The factor buys (goes long) a diversified portfolio of small stocks.
              * It sells (goes short) a diversified portfolio of big stocks.
                
          * What it means:
              * A positive SMB return means small-cap stocks outperformed large-cap stocks for that period.
              * A portfolio with a positive SMB beta is tilted toward small-cap stocks.


    * **HML (High Minus Low):** The value factor.

        * What it represents: This factor captures the value premium. This is the historical observation that "value" stocks (which appear cheap) tend to outperform "growth" stocks (which appear expensive).
          
         * How it's built:
             * Firms are sorted based on their Book-to-Market (B/M) ratio (the accounting value of the company divided by its stock market value).
             * A High B/M ratio means the stock is "cheap" (a value stock).
             * A Low B/M ratio means the stock is "expensive" (a growth stock).
             * The factor buys (goes long) a portfolio of High B/M (value) stocks.
             * It sells (goes short) a portfolio of Low B/M (growth) stocks.
           
        * What it means:
            * A positive HML return means value stocks outperformed growth stocks for that period.
            * A portfolio with a positive HML beta is tilted toward value stocks. A negative beta means it's tilted toward growth stocks.


    * **RMW (Robust Minus Weak):** The profitability factor.
 
        * What it represents: This factor captures the profitability premium. This is the observation that firms with high (robust) profitability tend to have higher average returns than firms with low (weak) profitability.
          
        * How it's built:
     
            * Firms are sorted based on their Operating Profitability (OP), which the paper measures as revenues minus costs, all divided by book equity.
            * The factor buys (goes long) a portfolio of Robust Profitability stocks.
            * It sells (goes short) a portfolio of Weak Profitability stocks.
              
        * What it means:
            * A positive RMW return means high-profitability firms outperformed low-profitability firms.
            * A portfolio with a positive RMW beta is tilted toward "quality" or high-profitability firms.

     
    * **CMA (Conservative Minus Aggressive):** The investment factor.

        * What it represents: This factor captures the investment premium. This is based on the theory and observation that companies that invest conservatively (grow their assets slowly) tend to have higher returns than firms that invest aggressively (grow their assets very quickly).
 
        * How it's built:
            * Firms are sorted based on their investment rate, which the paper measures as the year-over-year growth in total assets.
            * The factor buys (goes long) a portfolio of low-investment (Conservative) firms.
            * It sells (goes short) a portfolio of high-investment (Aggressive) firms.
              
        * What it means:
            * A positive CMA return means low-investment firms outperformed high-investment firms.
            * A portfolio with a positive CMA beta is tilted toward firms that are more conservative with their asset growth.

## 3. Replication Analysis

### (a) Model Specification: 

## (a) Model Specification

We estimate the Fama–French Five-Factor Model, which explains the excess return of a portfolio using five systematic risk factors. The regression equation is:

$$
R_{it} - R_{ft}
= \alpha_i
+ b_i \, (R_{Mt} - R_{ft})
+ s_i \, SMB_t
+ h_i \, HML_t
+ r_i \, RMW_t
+ c_i \, CMA_t
+ \varepsilon_{it}.
$$


where:

- **$R_{it} - R_{ft}$**  
  Portfolio *i*’s **excess return** at time *t*.  
  This is the dependent variable of the regression.

- **$\alpha$**  
  The portfolio’s **abnormal return** (intercept).  
  If the model prices the portfolio correctly, $\alpha_i$ should be statistically indistinguishable from zero.

- **$R_{Mt} - R_{ft}$** (MKT-RF)  
  The **market excess return**.  
  Captures the portfolio’s exposure to overall market risk.

- **$SMB_t$**  
  “Small Minus Big.”  
  The **size factor**, measuring the excess return of small firms over large firms.

- **$HML_t$**  
  “High Minus Low.”  
  The **value factor**, measuring the return spread between high book-to-market (value) and low book-to-market (growth) firms.

- **$RMW_t$**  
  “Robust Minus Weak.”  
  The **profitability factor**, capturing the return difference between firms with robust vs. weak operating profitability.

- **$CMA_t$**  
  “Conservative Minus Aggressive.”  
  The **investment factor**, measuring the return spread between firms that invest conservatively and those that invest aggressively.

- **$\varepsilon_{it}$**  
  Regression error term.

 **Coefficients**
- $b_i$: loading on the market factor  
- $s_i$: loading on the size factor  
- $h_i$: loading on the value factor  
- $r_i$: loading on the profitability factor  
- $c_i$: loading on the investment factor
  

These coefficients measure how sensitively the portfolio’s excess return responds to movements in each factor. The model is estimated using monthly data and Newey–West (HAC) standard errors to correct for potential autocorrelation and heteroskedasticity.


**Loading the Model:**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats

factors = pd.read_csv('5f_data.csv', skiprows=4)
factors = factors[pd.to_numeric(factors['Unnamed: 0'], errors='coerce').notna()]
factors = factors.rename(columns={'Unnamed: 0': 'date', 'Mkt-RF': 'MKTRF'})
factors = factors[['date', 'MKTRF', 'SMB', 'HML', 'RMW', 'CMA', 'RF']]
factors.iloc[:, 1:] = factors.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

port = pd.read_csv('25_portfolios_5x5.csv', skiprows=15)
port = port.rename(columns={'Unnamed: 0': 'date'})
port = port[pd.to_numeric(port['date'], errors='coerce').notna()]
port = port[:-1]
port.iloc[:, 1:] = port.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

factors = factors.drop_duplicates(subset='date')
port = port.drop_duplicates(subset='date')

data = pd.merge(port, factors, on='date', how='inner')
````

### (b) Estimation Results:

**Regression of Small, Low Book-to-Market Portfolios**

````python
cols = ['SMALL LoBM', 'RF', 'MKTRF', 'SMB', 'HML', 'RMW', 'CMA']
data[cols] = data[cols].apply(pd.to_numeric, errors='coerce')

data['excess'] = data['SMALL LoBM'] - data['RF']
data['excess'] = pd.to_numeric(data['excess'], errors='coerce')

reg = smf.ols('excess ~ MKTRF + SMB + HML + RMW + CMA',
              data=data).fit(cov_type='HAC', cov_kwds={'maxlags':12})

reg.summary()
````

````python
factor_cols = ['MKTRF', 'SMB', 'HML', 'RMW', 'CMA', 'RF']
portfolio_cols = [c for c in data.columns if c not in factor_cols + ['date']]

results = []

for port in portfolio_cols:

    excess = pd.to_numeric(data[port], errors='coerce') - pd.to_numeric(data['RF'], errors='coerce')
    data['excess'] = excess


    reg = smf.ols('excess ~ MKTRF + SMB + HML + RMW + CMA',
                  data=data).fit(cov_type='HAC', cov_kwds={'maxlags':12})

    results.append({
        'Portfolio': port,
        'Alpha': reg.params['Intercept'],
        't(Alpha)': reg.tvalues['Intercept'],
        'b_MKT': reg.params['MKTRF'],
        't(MKT)': reg.tvalues['MKTRF'],
        'b_SMB': reg.params['SMB'],
        't(SMB)': reg.tvalues['SMB'],
        'b_HML': reg.params['HML'],
        't(HML)': reg.tvalues['HML'],
        'b_RMW': reg.params['RMW'],
        't(RMW)': reg.tvalues['RMW'],
        'b_CMA': reg.params['CMA'],
        't(CMA)': reg.tvalues['CMA'],
        'R2': reg.rsquared
    })


results_df = pd.DataFrame(results)
results_df = results_df.set_index('Portfolio')
results_df
````
